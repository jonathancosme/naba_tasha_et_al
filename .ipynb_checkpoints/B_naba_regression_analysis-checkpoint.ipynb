### GPA

+ major GPA
+ overall GPA  

and calculated the entry z-score for each column

```python
major_gpa_col = 'GPA for your Major (For calculating instructions go to: "http://www.back2college.com/gpa.htm")'
overall_gpa_col = 'Overall GPA (For calculating instructions go to: "http://www.back2college.com/gpa.htm")'
```

```python
major_gpa_series = df[major_gpa_col].copy()
major_gpa_series.name = 'major_gpa'
overall_gpa_series = df[overall_gpa_col].copy()
overall_gpa_series.name = 'overall_gpa'
gpa_df = pd.concat([major_gpa_series, overall_gpa_series], axis=1)
gpa_df
```

```python
# # scale the columns by the maximum of the column (4.0 for both)
# gpa_df = gpa_df / gpa_df.max()

# scale the columns to z scores
gpa_df = (gpa_df - gpa_df.mean()) / gpa_df.std()

gpa_df
```

## target col
for column indicating if award was received:
1. replaced missing values with 'No'
2. converted column to binary (1 for "Yes 0 for "No")

```python
target_col = 'Recipient'
```

```python
df[target_col].unique()
```

```python
target_df = df[[target_col]].copy()
target_df[target_col] = ['No' if type(x) != str else x for x in target_df[target_col]]
target_df[target_col] = target_df[target_col].str.replace('No', '0')
target_df[target_col] = target_df[target_col].str.replace('Yes', '1')
target_df = target_df.astype(int)
target_df.value_counts()
```

```python
target_df
```

## make inputs DF

```python
inputs_df = pd.concat(
    [gender_df,
     clasf_one_hot_df,
     major_enc_df,
     major_2_enc_df,
     is_was_leader_df,
     did_acap_df,
     work_enc_df,
     did_cpa_rev_df,
     gpa_df,
     extra_curr_df,
     honors_df,
     paras_df
    ],
    axis=1,
)
inputs_df
```

```python
inputs_df.to_csv(reg_inputs_filename, index=False)
```

## Logit model 

```python
X = sm.add_constant(inputs_df.astype(float))
logit_reg = sm.Logit(
    target_df.astype(float),
    X,
#).fit(maxiter=1000, method='bfgs') # bfgs nm
).fit_regularized(maxiter=1000) # bfgs nm
logit_reg.summary()
```

```python
print(f"variables with statistical significance with alpha of {p_alpha} (p-values less than this):")
significant_var_coeffs = logit_reg.params[logit_reg.pvalues < p_alpha]
significant_var_coeffs
```

```python
significant_var_coeffs.reset_index().to_csv(significant_var_coeffs_filename, index=False)
```

```python
print(f"ODDS RATIOS of variables with statistical significance with alpha of {p_alpha} (p-values less than this):")
significant_var_odds_ratios = np.exp(significant_var_coeffs).sort_values()
significant_var_odds_ratios
```

```python
print("variables with negative effects and their odds ratios:")
significant_var_odds_ratios[significant_var_odds_ratios < 1 ]
```
